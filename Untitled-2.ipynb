```python
# Bài 1: Tính giai thừa
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Ví dụ: Tính giai thừa của 5
print("Giai thừa của 5 là:", factorial(5))

# Tính giai thừa của 10
print("Giai thừa của 10 là:", factorial(10))
```

```python
# Bài 2: Tính giá trị trung bình của dãy số
def average(numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)

# Ví dụ: Trung bình của dãy [1, 2, 3, 4, 5]
numbers = [1, 2, 3, 4, 5]
print("Trung bình của dãy", numbers, "là:", average(numbers))

# Trung bình của dãy số từ dữ liệu tài chính (ví dụ: Net_Income)
import pandas as pd
df = pd.read_csv('Financial Statement Anomaly Dataset.csv')
net_income_avg = average(df['Net_Income'].tolist())
print("Trung bình Net_Income từ dữ liệu là:", net_income_avg)
```

```python
# Bài 3: Tính lợi nhuận sau 12 tháng với số tiền ban đầu + tỷ lệ lãi suất
def compound_interest(principal, rate, time):
    # Công thức lãi suất kép: A = P * (1 + r)^t
    return principal * (1 + rate) ** time

# Ví dụ: Số tiền ban đầu 1000, lãi suất 5% (0.05), sau 12 tháng
principal = 1000
rate = 0.05  # 5%
time = 12  # tháng
profit = compound_interest(principal, rate, time) - principal
print(f"Lợi nhuận sau {time} tháng với số tiền ban đầu {principal} và lãi suất {rate*100}% là: {profit:.2f}")

# Áp dụng với dữ liệu tài chính: Sử dụng Total_Assets trung bình làm principal, giả sử lãi suất từ Return_on_Assets
avg_assets = df['Total_Assets'].mean()
avg_roa = df['Return_on_Assets'].mean()
profit_financial = compound_interest(avg_assets, avg_roa, 12) - avg_assets
print(f"Lợi nhuận giả định từ dữ liệu tài chính sau 12 tháng: {profit_financial:.2f}")
```
