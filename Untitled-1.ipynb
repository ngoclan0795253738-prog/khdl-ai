```python
# Khám phá cấu trúc dữ liệu
print("Thông tin về dữ liệu:")
df.info()

print("\nThống kê mô tả:")
print(df.describe())

print("\nKiểm tra giá trị thiếu:")
print(df.isnull().sum())

print("\nTên các cột:")
print(df.columns.tolist())
```

```python
# Xem tên cột và kích thước
print("Tên các cột:")
print(df.columns.tolist())
print(f"\nKích thước dữ liệu: {df.shape}")

# Kiểm tra kiểu dữ liệu
print("\nKiểu dữ liệu của các cột:")
print(df.dtypes)
```

```python
# Kiểm tra giá trị của Financial_Status
print("Giá trị unique trong Financial_Status:")
print(df['Financial_Status'].unique())
print("\nSố lượng mỗi giá trị:")
print(df['Financial_Status'].value_counts())
```

```python
# Kiểm tra giá trị thiếu
print("Số lượng giá trị thiếu trong mỗi cột:")
missing_values = df.isnull().sum()
print(missing_values)
print(f"\nTổng giá trị thiếu: {missing_values.sum()}")
```

```python
# Phân tích theo Financial_Status
print("Thống kê theo nhóm Financial_Status:")
grouped = df.groupby('Financial_Status').describe().T
print(grouped)
```

```python
# Trực quan hóa phân phối của một số chỉ số theo Financial_Status
import matplotlib.pyplot as plt
import seaborn as sns

# Chọn một số cột quan trọng
cols_to_plot = ['Net_Income', 'Current_Ratio', 'Debt_to_Equity', 'Return_on_Assets']

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for i, col in enumerate(cols_to_plot):
    ax = axes[i//2, i%2]
    sns.boxplot(x='Financial_Status', y=col, data=df, ax=ax)
    ax.set_title(f'Boxplot of {col} by Financial_Status')

plt.tight_layout()
plt.show()
```

```python
# Tính ma trận tương quan
numeric_cols = df.select_dtypes(include=[float]).columns
corr_matrix = df[numeric_cols].corr()

# Vẽ heatmap tương quan
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Ma trận tương quan giữa các chỉ số tài chính')
plt.show()
```

# Giải thích Chi tiết Kết quả Biểu đồ

## 1. Boxplots: Phân phối Các Chỉ số Tài chính Theo Nhóm Financial_Status

Boxplot là một cách trực quan để xem phân phối dữ liệu, bao gồm:
- **Hộp (box):** Chứa 50% dữ liệu giữa quartile thứ 1 (Q1) và thứ 3 (Q3). Đường giữa là median (Q2).
- **Ria (whiskers):** Mở rộng đến giá trị tối đa/min trong phạm vi 1.5 lần khoảng cách giữa Q1 và Q3.
- **Điểm ngoài (outliers):** Các giá trị nằm ngoài ria, được đánh dấu bằng dấu chấm.

### Giải thích Chi tiết Mỗi Biểu đồ:

#### a. Net_Income (Lợi nhuận Ròng)
- **Normal:** Median gần 0, phân phối cân đối, ít outliers. Cho thấy lợi nhuận ổn định.
- **High Risk:** Median thấp hơn (khoảng -0.05), nhiều outliers ở phía âm. Gợi ý rủi ro cao với lợi nhuận kém.
- **Anomaly:** Median thấp nhất (khoảng -0.02), phân phối rộng hơn, nhiều outliers. Bất thường rõ rệt với lợi nhuận biến động và thường âm.

#### b. Current_Ratio (Tỷ lệ Thanh toán Hiện tại)
- **Normal:** Median gần 0, phân phối hẹp, ít outliers. Tài chính lành mạnh, khả năng thanh toán tốt.
- **High Risk:** Median thấp hơn, nhiều outliers âm. Khó khăn trong thanh toán ngắn hạn.
- **Anomaly:** Phân phối rộng, median thấp, nhiều outliers. Bất thường với tỷ lệ thanh toán kém.

#### c. Debt_to_Equity (Tỷ lệ Nợ trên Vốn Chủ sở hữu)
- **Normal:** Median gần 0, cân đối.
- **High Risk:** Median cao hơn, cho thấy nợ nhiều hơn vốn.
- **Anomaly:** Phân phối lệch, nhiều outliers cao. Bất thường với cấu trúc vốn không bền vững.

#### d. Return_on_Assets (ROA - Hiệu suất Tài sản)
- **Normal:** Median dương nhẹ, phân phối tốt.
- **High Risk:** Median âm, hiệu suất kém.
- **Anomaly:** Median thấp nhất, nhiều outliers âm. Bất thường với hiệu suất tài sản rất kém.

**Tóm lại:** Các nhóm Anomaly và High Risk có xu hướng có giá trị thấp hơn ở các chỉ số lợi nhuận và hiệu suất, với nhiều biến động hơn (outliers). Điều này giúp phân biệt chúng với Normal.

## 2. Heatmap: Ma trận Tương quan Giữa Các Chỉ số Tài chính

Heatmap hiển thị mức độ mối quan hệ giữa các biến:
- **Màu đỏ đậm:** Tương quan dương cao (khoảng 0.8-1.0) - hai biến tăng/giảm cùng nhau.
- **Màu xanh đậm:** Tương quan âm cao (-0.8 đến -1.0) - một biến tăng thì biến kia giảm.
- **Màu vàng/trắng:** Tương quan thấp (gần 0) - không liên quan.

### Các Mối quan hệ Quan trọng:
- **Revenue và Operating_Expenses:** Tương quan cao dương (~0.85). Logic vì chi phí vận hành thường tỷ lệ với doanh thu.
- **Net_Income và Return_on_Assets/Return_on_Equity:** Cao dương (~0.8-0.9). Lợi nhuận cao dẫn đến ROA/ROE tốt.
- **Total_Assets và Total_Liabilities:** Cao dương (~0.8). Tài sản và nợ thường đi đôi.
- **Cash_Flow_Operating và Net_Income:** Cao dương (~0.9). Dòng tiền từ hoạt động phản ánh lợi nhuận.
- **Debt_to_Equity và Current_Ratio:** Âm (~ -0.8). Nợ cao thường làm giảm tỷ lệ thanh toán.

**Tóm lại:** Heatmap cho thấy các chỉ số tài chính liên kết chặt chẽ, đặc biệt giữa doanh thu, chi phí và lợi nhuận. Điều này hữu ích để hiểu tác động chéo và phát hiện bất thường khi một chỉ số thay đổi bất thường so với các chỉ số liên quan.

## Đánh giá Rủi ro và Độ tin cậy từ Biểu đồ

### Rủi ro
- **Anomaly và High Risk** có phân phối rộng hơn và nhiều outliers hơn, đặc biệt ở các chỉ số lợi nhuận và hiệu suất như `Net_Income`, `Return_on_Assets`, và `Current_Ratio`. Điều này cho thấy các mẫu này có mức độ biến động cao, nguy cơ thua lỗ và thanh khoản kém.
- **High Risk** có median và phần lớn giá trị thấp hơn so với `Normal`, nên đây là nhóm rủi ro cao hơn, dễ gặp khó khăn vì nợ lớn và lợi nhuận thấp.
- **Anomaly** thường thể hiện các giá trị bất thường rõ rệt hơn, có thể là dấu hiệu bất ổn về báo cáo tài chính hoặc số liệu không ổn định.

### Độ tin cậy
- **Normal** có phân phối phần lớn gọn, ít outliers, cho thấy dữ liệu ổn định và tin cậy hơn. Các chỉ số trong nhóm này phù hợp với mô hình tài chính lành mạnh.
- **Heatmap tương quan** cho thấy các chỉ số chính như `Revenue`, `Operating_Expenses`, `Net_Income`, và `Cash_Flow_Operating` có mối quan hệ chặt chẽ. Khi các biến có tương quan mạnh, mô hình dự đoán sẽ có độ tin cậy tốt hơn nếu dữ liệu không bị nhiễu quá nhiều.
- Tuy nhiên, **đối với các nhóm Anomaly/High Risk**, độ tin cậy của dự đoán giảm nếu chúng ta chỉ dùng một vài chỉ số vì phân phối biến động mạnh và nhiều outliers. Cần xem xét thêm nhiều chỉ số và dùng kỹ thuật robust để giảm ảnh hưởng của dữ liệu bất thường.

### Kết luận
- Nhóm `Normal`: độ tin cậy cao, rủi ro thấp.
- Nhóm `High Risk`: rủi ro cao, đặc biệt khi `Debt_to_Equity` tăng và `Current_Ratio` giảm.
- Nhóm `Anomaly`: biểu hiện rủi ro không ổn định, cần kiểm tra thêm để xác định xem đó là lỗi dữ liệu hay tình trạng tài chính bất thường.
